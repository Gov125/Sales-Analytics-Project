DATA LOADED


```python
import pandas as pd
df_merge=pd.read_csv("Orders.csv").merge(pd.read_csv("Details.csv"), on='Order ID')

df_merge.head()
```

DATA CLEANING

```python
# Check missing values
print("Missing values :")
print(df_merge.isnull().sum())
```

```python
#Dataset overview
print("\n Summary statistics :")
print(df_merge.describe())
```

```python
print("\n Data info:")
print(df_merge.info())
```

```python
#Convert Order Date from object to datetime
df_merge["Order Date"] = pd.to_datetime(df_merge["Order Date"],dayfirst=True)
```

```python
#Check for duplicates
print("\nNumber of duplicate rows :")
print(df_merge.duplicated().sum())
```

```python
#Check for negative values in Amount, Profit, Quantity
print("\nNegative values in Amount:")
print(df_merge[df_merge["Amount"] < 0])
print("\nNegative values in Profit:")
print(df_merge[df_merge["Profit"] < 0])
print("\nNegative values in Quantity:")
print(df_merge[df_merge["Quantity"] < 0])
```

```python
#Chweck for unique values in categorical columns
print("\nUnique values in State:")
print(df_merge["State"].unique())
print("\nUnique values in Category:")
print(df_merge["Category"].unique())
print("\nUnique values in Sub-Category:")
print(df_merge["Sub-Category"].unique())
print("\nUnique values in Payment Mode:")
print(df_merge["PaymentMode"].unique())
print("\nUnique values in City:")
print(df_merge["City"].unique())
```

DATA VISUALIZATION

```python
#Sales by category
import matplotlib.pyplot as plt
df_merge.groupby("Category")["Amount"].sum().plot(kind="bar", figsize=(12,6))
plt.title("sales by category")
plt.xlabel("Category") 
plt.ylabel(" Amount")
plt.show()
```

```python
#Profit by category
df_merge.groupby("Category")["Profit"].sum().plot(kind="bar",figsize=(10,6))
plt.title("profit by category")
plt.xlabel("Category")
plt.ylabel("Profit")
plt.show()
```
